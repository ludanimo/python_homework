1. Создайте массив 3x3, состоящий из True. Напишите минимум два варианта решения


```python
import numpy as np
```

```python
array_1 = np.ones((3, 3), dtype=bool)
print(array_1)
```

```python
array_2 = np.array([True] * 9).reshape(3, 3)
print(array_2)
```

2. Замените в массиве все нечетные положительные числа на -1

```python
matrix_1 = np.arange(20).reshape((4, 5))
print(matrix_1)

matrix_2 = matrix_1.copy()
matrix_2[matrix_1 % 2 != 0] = - 1
print(matrix_2)
```

3. Найдите все числа, встречающиеся в двух массивах

```python
numbers_elem = np.intersect1d(matrix_1, matrix_2)
print(numbers_elem)
```

4. Дана матрица 3х3. Поменяйте местами 1 и 2 столбцы матрицы.

```python
matrix_3 = np.arange(9).reshape(3, 3)
print(matrix_3)

matrix_3[:, [0, 1]] = matrix_3[:, [1, 0]]
print(matrix_3)
```

5. В данном массиве замените все числа, меньшие 0 на 0, а большие 255 на 255. Напишите минимум два варианта решения

```python
array_3 = np.ones((6, 7))
array_3[::2, 1::2] = 777
array_3[1::2, ::2] = -1
print(array_3)
```

```python
array_3 = np.clip(array_3, 0, 255)
print(array_3)
```

```python
array_3[array_3 < 0] = 0
array_3[array_3 > 255] = 255
print(array_3)
```

6. Преобразуйте массив массивов в одномерный массив.

```python
one_dim = array_3.flatten()
print(one_dim)
```

7. Найдите максимум в каждой строке двумерного массива

```python
max = np.max(array_3, axis=1)
print(max)
```

8. Удалите все np.nan из массива.
*Для того, чтобы в тестовом массиве появился nan используйте np.nan или деление 0. на 0.*

```python
array_3[array_3 == 255.] = np.nan
print(array_3)
```

```python
no_nan = array_3[~np.isnan(array_3)].copy()
print(no_nan)
```

9. Загрузите изображение, сохраненное в формате .npy lena.npy или любое другое изображение в формате png.
Используйте известные вам способы модификации numpy массива для того, чтобы изменить изображение. Отсматривайте результаты, используя функцию для отрисовки изображений plt.imshow().

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open("wall.png")
```

```python
img = np.array(img)
```

```python
np.save("wall.npy", img)
```

```python
img = np.load("wall.npy")
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
cropped_image = img[50:300, 100:400] # Обрезка
plt.imshow(cropped_image)
```

```python
brightened_image = np.clip(img + 0.7, 0, 1)  # Увеличиваем каждый канал на 0.7 и ограничиваем значения до [0, 1]
plt.imshow(brightened_image)
plt.title("Ув. Яркость")
```

```python
saturated_image = np.clip(img * 20.0, 0, 1) # Увеличиваем насыщенность в двадцать раз
plt.imshow(saturated_image)
plt.title("Ув. Насыщенность")
```
